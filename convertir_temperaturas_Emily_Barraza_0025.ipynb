### 1. Fórmulas de Conversión

Para realizar las conversiones entre escalas térmicas, utilizamos las siguientes expresiones matemáticas:

* **De Grados Celsius ($^{\circ}C$) a Grados Fahrenheit ($^{\circ}F$):**
    $$^{\circ}F = (^{\circ}C \times \frac{9}{5}) + 32$$
    *(También se puede usar $1.8$ en lugar de $\frac{9}{5}$)*.

* **De Grados Fahrenheit ($^{\circ}F$) a Grados Celsius ($^{\circ}C$):**
    $$^{\circ}C = \frac{^{\circ}F - 32}{1.8}$$
    *(O bien, $^{\circ}C = (^{\circ}F - 32) \times \frac{5}{9}$)*.

### 2. Ejemplos Resueltos Paso a Paso

**Ejemplo 1: Convertir $25^{\circ}C$ a $^{\circ}F$ (Temperatura ambiente)**
1.  **Identificar datos:** Celsius = $25$.
2.  **Aplicar fórmula:** $^{\circ}F = (25 \times 1.8) + 32$.
3.  **Multiplicación:** $25 \times 1.8 = 45$.
4.  **Suma:** $45 + 32 = 77$.
5.  **Resultado:** $25^{\circ}C$ equivalen a **$77^{\circ}F$**.

**Ejemplo 2: Convertir $98.6^{\circ}F$ a $^{\circ}C$ (Temperatura corporal promedio)**
1.  **Identificar datos:** Fahrenheit = $98.6$.
2.  **Aplicar fórmula:** $^{\circ}C = (98.6 - 32) / 1.8$.
3.  **Resta:** $98.6 - 32 = 66.6$.
4.  **División:** $66.6 / 1.8 = 37$.
5.  **Resultado:** $98.6^{\circ}F$ equivalen a **$37^{\circ}C$**.

### 3. Código en Python

Aquí tienes el programa interactivo siguiendo todas tus especificaciones:

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    try:
        celsius = float(input("\nIngrese los grados Celsius a convertir: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"Resultado: {celsius}°C es igual a {fahrenheit:.2f}°F")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    try:
        fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"Resultado: {fahrenheit}°F es igual a {celsius:.2f}°C")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def menu():
    """Función principal con el menú interactivo"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Llamada a la función principal para iniciar el programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
1.  **`while True`**: Crea un bucle infinito para que el menú se repita hasta que el usuario decida salir explícitamente.
2.  **`input()`**: Captura la opción del menú y los valores numéricos de temperatura.
3.  **`if/elif`**: Estructura de control que decide qué función ejecutar según la entrada del usuario.
4.  **`try/except`**: Se añadió para evitar que el programa se cierre si el usuario escribe letras en lugar de números al ingresar la temperatura.
5.  **`break`**: Rompe el ciclo `while` inmediatamente cuando se selecciona la opción 3.

````python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    try:
        celsius = float(input("\nIngrese los grados Celsius a convertir: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"Resultado: {celsius}°C es igual a {fahrenheit:.2f}°F")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    try:
        fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"Resultado: {fahrenheit}°F es igual a {celsius:.2f}°C")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def menu():
    """Función principal con el menú interactivo"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡Emily Barraza NL = 7!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Llamada a la función principal para iniciar el programa
if __name__ == "__main__":
    menu()
````
